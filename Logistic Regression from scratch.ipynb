```python
class LogisticRegression:
    
    def __init__(self, lr=0.001, n_it=1000):
        self.lr=lr
        self.n_it = n_it
        self.weights = None
        self.bias = None
        
    def fit(self, X , y): #X -> mxn Y-> 1xm
        #Initialize:
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        #Gradient Descent Application:
        for i in range(self.n_it):
            y_lin = np.dot(X, self.weights + self.bias)
            y_pred = self._sigmoid(y_lin)
            
            #Derivatives Calculations:
            dw = 2*(1/n_samples)*np.dot(X.T, y_pred-y)
            db = 2*(1/n_samples)*np.sum(y_pred-y)
            
            #Refining the weights:
            self.weights -= self.lr*dw
            self.bias -= self.lr*db
                    
    def predict(self, X):
        y_lin = np.dot(X, self.weights + self.bias)
        y_pred = self._sigmoid(y_lin)
        y_pred_cls = [1 if i>0.5 else 0 for i in y_pred]
        return y_pred_cls
        
        
    def _sigmoid(self, x):
        return 1/(1+np.exp(-x))
```

```python
regressor = LogisticRegression(lr=0.001, n_it=1000)
regressor.fit(X_train, y_train)
predictions = regressor.predict(X_test)
```

```python
accuracy_score(y_test, predictions)
```

```python
confusion_matrix(y_test, predictions)
```

```python
f1_score(y_test, predictions)
```
